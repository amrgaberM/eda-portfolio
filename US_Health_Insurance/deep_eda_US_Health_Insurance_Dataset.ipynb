```python
from scipy.stats import lognorm

# Filter the DataFrame for smokers and non-smokers
smokers_df = df[df['smoker'] == 'yes'].copy()
non_smokers_df = df[df['smoker'] == 'no'].copy()

# Fit log-normal distribution to charges for smokers
s_smokers, loc_smokers, scale_smokers = lognorm.fit(smokers_df['charges'], floc=0)

# Fit log-normal distribution to charges for non-smokers
s_non_smokers, loc_non_smokers, scale_non_smokers = lognorm.fit(non_smokers_df['charges'], floc=0)

# --- Plotting for Smokers ---
# Sort the charges for smokers
sorted_charges_smokers = np.sort(smokers_df['charges'])
# Calculate the empirical CDF for smokers
cdf_charges_smokers = np.arange(1, len(sorted_charges_smokers) + 1) / len(sorted_charges_smokers)

plt.figure(figsize=(10, 6))
plt.plot(sorted_charges_smokers, cdf_charges_smokers, label='Empirical CDF (Smokers)')

# Plot the theoretical log-normal CDF for smokers
x_smokers = np.linspace(smokers_df['charges'].min(), smokers_df['charges'].max(), 100)
plt.plot(x_smokers, lognorm.cdf(x_smokers, s_smokers, loc=loc_smokers, scale=scale_smokers), 'r-', lw=2, label='Fitted Log-normal CDF (Smokers)')

plt.title('CDF of Charges with Fitted Log-normal Distribution (Smokers)')
plt.xlabel('Charges')
plt.ylabel('CDF')
plt.legend()
plt.grid(True)
plt.show()

# --- Plotting for Non-Smokers ---
# Sort the charges for non-smokers
sorted_charges_non_smokers = np.sort(non_smokers_df['charges'])
# Calculate the empirical CDF for non-smokers
cdf_charges_non_smokers = np.arange(1, len(sorted_charges_non_smokers) + 1) / len(sorted_charges_non_smokers)

plt.figure(figsize=(10, 6))
plt.plot(sorted_charges_non_smokers, cdf_charges_non_smokers, label='Empirical CDF (Non-Smokers)')

# Plot the theoretical log-normal CDF for non-smokers
x_non_smokers = np.linspace(non_smokers_df['charges'].min(), non_smokers_df['charges'].max(), 100)
plt.plot(x_non_smokers, lognorm.cdf(x_non_smokers, s_non_smokers, loc=loc_non_smokers, scale=scale_non_smokers), 'r-', lw=2, label='Fitted Log-normal CDF (Non-Smokers)')

plt.title('CDF of Charges with Fitted Log-normal Distribution (Non-Smokers)')
plt.xlabel('Charges')
plt.ylabel('CDF')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Filter the DataFrame for smokers
smokers_df = df[df['smoker'] == 'yes'].copy()

# Apply log transformation to charges for smokers (if not already done)
if 'log_charges' not in smokers_df.columns:
    smokers_df['log_charges'] = np.log(smokers_df['charges'])

# Calculate the mean and standard deviation of the log-transformed charges for smokers
mean_log_charges_smokers = smokers_df['log_charges'].mean()
std_log_charges_smokers = smokers_df['log_charges'].std()

print(f"Mean of log-transformed charges for smokers: {mean_log_charges_smokers:.4f}")
print(f"Standard deviation of log-transformed charges for smokers: {std_log_charges_smokers:.4f}")
```

```python
# Calculate the pooled standard deviation
pooled_std = np.sqrt(((len(smokers_df) - 1) * std_log_charges_smokers**2 + (len(non_smokers_df) - 1) * std_log_charges_non_smokers**2) / (len(smokers_df) + len(non_smokers_df) - 2))

# Calculate Cohen's d
cohens_d = (mean_log_charges_smokers - mean_log_charges_non_smokers) / pooled_std

print(f"Cohen's d (effect size) for the difference in log-transformed charges between smokers and non-smokers: {cohens_d:.4f}")
```

```python
# Calculate the mean charges for smokers and non-smokers
mean_charges_smokers = df[df['smoker'] == 'yes']['charges'].mean()
mean_charges_non_smokers = df[df['smoker'] == 'no']['charges'].mean()

# Calculate the difference between the means
mean_difference = mean_charges_smokers - mean_charges_non_smokers

print(f"Mean charges for smokers: {mean_charges_smokers:.2f}")
print(f"Mean charges for non-smokers: {mean_charges_non_smokers:.2f}")
print(f"Difference in mean charges (Smokers - Non-smokers): {mean_difference:.2f}")
```

```python
# Combine the charges of smokers and non-smokers
combined_charges = df['charges']
smoker_labels = df['smoker']

# Calculate the observed difference in means (already calculated in a previous cell)
observed_mean_difference = df[df['smoker'] == 'yes']['charges'].mean() - df[df['smoker'] == 'no']['charges'].mean()

# Number of permutations
n_permutations = 10000

# Array to store permutation differences
permutation_diffs = np.zeros(n_permutations)

for i in range(n_permutations):
    # Permute the smoker labels
    permuted_labels = np.random.permutation(smoker_labels)

    # Calculate the mean charges for the permuted groups
    permuted_mean_smokers = combined_charges[permuted_labels == 'yes'].mean()
    permuted_mean_non_smokers = combined_charges[permuted_labels == 'no'].mean()

    # Calculate the difference in means for the permutation
    permutation_diffs[i] = permuted_mean_smokers - permuted_mean_non_smokers

# Calculate the p-value
# The p-value is the proportion of permutation differences that are as extreme as or more extreme than the observed difference
p_value = np.sum(np.abs(permutation_diffs) >= np.abs(observed_mean_difference)) / n_permutations

print(f"Observed mean difference: {observed_mean_difference:.4f}")
print(f"P-value from permutation test: {p_value:.4f}")

# Optional: Plot the distribution of permutation differences
plt.figure(figsize=(10, 6))
plt.hist(permutation_diffs, bins=50, density=True, alpha=0.6, label='Permutation Differences')
plt.axvline(observed_mean_difference, color='red', linestyle='--', label='Observed Difference')
plt.title('Distribution of Permutation Differences in Mean Charges')
plt.xlabel('Difference in Mean Charges')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
from scipy.optimize import curve_fit

# Define the linear function
def linear_func(x, a, b):
    return a * x + b

# Get the age and charges data
age_data = df['age']
charges_data = df['charges']

# Fit the linear function to the data using curve_fit (least squares)
params, covariance = curve_fit(linear_func, age_data, charges_data)

# Extract the fitted parameters (slope and intercept)
slope, intercept = params

print(f"Fitted line: charges = {slope:.2f} * age + {intercept:.2f}")

# Generate points for the fitted line
age_range = np.linspace(age_data.min(), age_data.max(), 100)
fitted_charges = linear_func(age_range, slope, intercept)

# Plot the original data and the fitted line
plt.figure(figsize=(10, 6))
plt.scatter(age_data, charges_data, alpha=0.5, label='Original Data')
plt.plot(age_range, fitted_charges, color='red', label='Fitted Line')
plt.title('Linear Fit (Least Squares) between Age and Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate the predicted charges based on the fitted line
predicted_charges = linear_func(age_data, slope, intercept)

# Calculate the residuals
residuals = charges_data - predicted_charges

# Plot a histogram of the residuals
plt.figure(figsize=(10, 6))
sns.histplot(residuals, kde=True)
plt.title('Distribution of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()

# Plot residuals vs. fitted values
plt.figure(figsize=(10, 6))
plt.scatter(predicted_charges, residuals, alpha=0.5)
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Residuals vs. Fitted Values')
plt.xlabel('Fitted Charges')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()

# Plot residuals vs. age
plt.figure(figsize=(10, 6))
plt.scatter(age_data, residuals, alpha=0.5)
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Residuals vs. Age')
plt.xlabel('Age')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Calculate the predicted charges based on the fitted line (from the previous cell)
# predicted_charges is already available from cell 2Y4mbT_eVqG0

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(charges_data, predicted_charges))

# Calculate R2
r2 = r2_score(charges_data, predicted_charges)

print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R2): {r2:.4f}")
```

```python
# Number of bootstrap samples
n_bootstrap_samples = 1000

# Array to store bootstrap sample means
bootstrap_means = np.zeros(n_bootstrap_samples)

for i in range(n_bootstrap_samples):
    # Resample the data with replacement
    bootstrap_sample = df['charges'].sample(frac=1, replace=True)

    # Calculate the mean of the bootstrap sample
    bootstrap_means[i] = bootstrap_sample.mean()

# Calculate the 95% confidence interval (2.5th and 97.5th percentiles)
confidence_interval = np.percentile(bootstrap_means, [2.5, 97.5])

print(f"Bootstrap 95% Confidence Interval for the mean of Charges: [{confidence_interval[0]:.2f}, {confidence_interval[1]:.2f}]")

# Optional: Plot the distribution of bootstrap means
plt.figure(figsize=(10, 6))
plt.hist(bootstrap_means, bins=50, density=True, alpha=0.6, label='Bootstrap Means')
plt.axvline(confidence_interval[0], color='red', linestyle='--', label='2.5th Percentile')
plt.axvline(confidence_interval[1], color='red', linestyle='--', label='97.5th Percentile')
plt.title('Distribution of Bootstrap Sample Means')
plt.xlabel('Mean Charges')
plt.ylabel('Density')
plt.legend()
plt.show()
```
